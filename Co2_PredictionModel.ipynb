```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```
df=pd.read_csv('/content/AQ/AirQualityUCI.csv', sep=';', decimal=',')
df.head()
```

```
df.replace(-200, np.nan, inplace=True)
```

```
features=df[[
        "PT08.S1(CO)",
    "PT08.S2(NMHC)",
    "PT08.S3(NOx)",
    "PT08.S4(NO2)",
    "PT08.S5(O3)",
    "T",
    "RH",
    "AH",
    "CO(GT)"
]]
display(features)
```

```
features.dropna(inplace=True)
```

```
x=features.drop("CO(GT)", axis=1)
y=features["CO(GT)"]
```

```
all_features=Pipeline(steps=[
  ('imputer',SimpleImputer(strategy='mean')),
  ('scaler', StandardScaler()),
]
)

preprocessor=ColumnTransformer(
    transformers=[
        ('all', all_features, x.columns)
    ]
)
```

```
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.2, random_state=30)
model.fit(x_train, y_train)
```

```
r2_score = model.score(x_test, y_test)
r2_score=r2_score*100
print(f"R² Score: {r2_score:.2f}%")
```

```
# Generate one random row of values (within approximate realistic ranges)
random_data = pd.DataFrame([{
    "PT08.S1(CO)": np.random.randint(500, 2000),
    "PT08.S2(NMHC)": np.random.randint(500, 2000),
    "PT08.S3(NOx)": np.random.randint(200, 2000),
    "PT08.S4(NO2)": np.random.randint(500, 2000),
    "PT08.S5(O3)": np.random.randint(500, 2000),
    "T": np.random.uniform(0, 35),       # °C
    "RH": np.random.uniform(10, 90),     # %
    "AH": np.random.uniform(0.5, 2.0)    # g/m³
}])

print("🔹 Random Input Data:")
print(random_data)

# Predict CO concentration
predicted_CO = model.predict(random_data)
print(f"\n Predicted CO(GT): {predicted_CO[0]:.4f} mg/m³")
```
